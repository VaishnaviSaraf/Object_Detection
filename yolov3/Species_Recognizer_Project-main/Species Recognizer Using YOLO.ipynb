```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
yolo = cv2.dnn.readNet("./yolov3.weights", "./yolov3.cfg")
```

```python
classes = []

with open("./coco.names", 'r') as f:
    classes = f.read().splitlines()
```

```python
classes
```

```python
len(classes)
```

```python
img = cv2.imread("./test1.jpg")
```

```python
blob = cv2.dnn.blobFromImage(img, 1/255, (320,320), (0,0,0), swapRB=True, crop=False)
```

```python
blob.shape
```

```python
# to print img:

i = blob[0].reshape(320,320,3)
plt.imshow(i)
```

```python
yolo.setInput(blob)
```

```python
output_layers_names = yolo.getUnconnectedOutLayersNames()
layeroutput = yolo.forward(output_layers_names)
```

```python
height, width, _ = img.shape
```

```python
height
```

```python
width
```

```python
boxes = []
confidences = []
class_ids = []

for output in layeroutput:
    for detection in output:
        score = detection[5:]
        class_id = np.argmax(score)
        confidence = score[class_id]
        
        if confidence > 0.7 :
            center_x = int(detection[0]*width)
            center_y = int(detection[1]*height)
            
            w = int(detection[2]*width)
            h = int(detection[3]*height)
            
            x = int(center_x - w/2)
            y = int(center_y - h/2)
            
            boxes.append([x,y,w,h])
            confidences.append(float(confidence))
            class_ids.append(class_id)
```

```python
len(boxes)
```

```python
indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
```

```python
font = cv2.FONT_HERSHEY_PLAIN
colors = np.random.uniform(0, 255, size = (len(boxes), 3) )
```

```python
for i in indexes.flatten():
    x,y,w,h = boxes[i]
    
    label = str(classes[class_ids[i]])
    confi = str(round(confidences[i], 2))
    color = colors[i]
    
    cv2.rectangle(img, (x,y), (x+w, y+h), color, 3)
    cv2.putText(img, label+" "+confi, (x,y+20), font, 2, (255,255,255), 3)
# for i in indexes.flatten():
#     x, y, w, h = boxes[i]
    
#     label = str(classes[class_ids[i]])
#     confi = str(round(confidences[i], 2))
#     color = (0, 0, 0)  # Set color to black
    
#     cv2.rectangle(img, (x, y), (x + w, y + h), color, 1)
#     cv2.putText(img, label + " " + confi, (x, y + 20), font, 10, (0, 0, 0), 1)  # Set text color to black
```
